```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
pso_df = pd.read_csv("../results/pso_history.csv")
ga_df = pd.read_csv("../results/ga_history.csv")
sa_df = pd.read_csv("../results/sa_history.csv")
tabu_df = pd.read_csv("../results/tabu_history.csv")

# Plot
plt.figure(figsize=(12,7))
plt.plot(pso_df['iteration'], pso_df['best_score'], label='PSO', marker='o', linewidth=2)
plt.plot(ga_df['iteration'], ga_df['best_score'], label='GA', marker='x', linewidth=2)
plt.plot(sa_df['iteration'], sa_df['best_score'], label='SA', marker='s', linewidth=2)
plt.plot(tabu_df['iteration'], tabu_df['best_score'], label='Tabu Search', marker='^', linewidth=2)

plt.title("Algorithm Convergence Comparison (TinyBERT)", fontsize=14, fontweight='bold')
plt.xlabel("Iteration / Generation", fontsize=12)
plt.ylabel("Validation Accuracy", fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print Analysis
print("="*60)
print("FINAL RESULTS - BEST ACCURACY")
print("="*60)
print(f"PSO:          {pso_df['best_score'].max():.4f}")
print(f"GA:           {ga_df['best_score'].max():.4f}")
print(f"SA:           {sa_df['best_score'].max():.4f}")
print(f"Tabu Search:  {tabu_df['best_score'].max():.4f}")
print("="*60)
```
